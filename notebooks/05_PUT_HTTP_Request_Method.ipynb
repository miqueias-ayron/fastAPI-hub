# PUT HTTP Request Method

O funcionamento do **PUT** vai ser um pouco parecido com o **POST** mas com a diferença de que, enquanto o POST é usado para **criação** de novos dados, o **PUT** **atualiza** dados que já existem. 

Apesar dessa diferença, o **PUT Request** aceita exatamente o mesmo corpo de requisição do **POST**.

```python
from fastapi import FastAPI, Body

app = FastAPI()

BOOKS = [
    {'title': 'Title One', 'author': 'Author One','category': 'science'},
    {'title': 'Title Two', 'author': 'Author Two','category': 'science'},
    {'title': 'Title Three', 'author': 'Author Three','category': 'history'},
    {'title': 'Title Four', 'author': 'Author Four','category': 'math'},
    {'title': 'Title Five', 'author': 'Author Five','category': 'math'},
    {'title': 'Title Six', 'author': 'Author Two', 'category': 'math'}
]

@app.put('/books/update_book')
async def update_book(updated_book=Body()):
    for i in range(len(BOOKS)):
        if BOOKS[i].get('title').casefold() == updated_book.get('title').casefold():
            BOOKS[i] == updated_book
```
